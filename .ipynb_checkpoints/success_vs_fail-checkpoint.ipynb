```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Таблица бронирований

```python
df=pd.read_csv('alldata.csv', encoding = 'UTF-8')
df
```

# 2. Группировка столбцов по успешным бронированиям

## По каналам бронирования (интернет, мобильное устройство)

```python
channel = df.groupby('booking_complete')['sales_channel'].value_counts(normalize=True).to_frame().reset_index()
channel
```

```python
plt.figure(figsize=(7,5))
plt.subplot(1, 1, 1)
sns.barplot(x='booking_complete', y='proportion', hue = 'sales_channel', data=channel, errorbar=None)
# channel.loc[0].plot.bar()
# plt.subplot(1, 2, 2)
# sns.barplot(x='sales_channel', y='proportion', data=channel, errorbar=None)
#channel.loc[1].plot.bar()
```

## По количеству пассажиров

```python
passengers=df.groupby('booking_complete')['num_passengers'].value_counts(normalize=True).to_frame().reset_index()
passengers
```

```python
plt.figure(figsize=(7,5))
plt.subplot(1, 1, 1)
sns.barplot(x='booking_complete', y='proportion', hue = 'num_passengers', data=passengers, errorbar=None)
```

## По типу поездки

```python
trip = df.groupby('booking_complete')['trip_type'].value_counts(normalize=True).to_frame().reset_index()
trip 
```

```python
plt.figure(figsize=(7,5))
plt.subplot(1, 1, 1)
sns.barplot(x='booking_complete', y='proportion', hue = 'trip_type', data=trip, errorbar=None)
```

## По времени вылета

```python
departure_hour = df.groupby('booking_complete')['flight_hour'].value_counts(normalize=True).to_frame().sort_values(['booking_complete','flight_hour']).reset_index()
departure_hour 
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1, 1, 1)
sns.barplot(x='booking_complete', y='proportion', hue = 'flight_hour', data=departure_hour, errorbar=None)
```

## По длительности полета

```python
duration_of_flight = df.groupby('booking_complete')['flight_duration'].value_counts(normalize=True).to_frame().sort_values(['booking_complete', 'flight_duration']).reset_index()
duration_of_flight
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1, 1, 1)
sns.barplot(x='booking_complete', y='proportion', hue = 'flight_duration', data=duration_of_flight, errorbar=None)
```

## По необходимости дополнительного багажа

```python
extra_baggage = df.groupby('booking_complete')['wants_extra_baggage'].value_counts(normalize=True).to_frame().reset_index()
extra_baggage
```

```python
plt.figure(figsize=(5,5))
plt.subplot(1, 1, 1)
sns.barplot(x='booking_complete', y='proportion', hue = 'wants_extra_baggage', data=extra_baggage, errorbar=None)
```

## По необходимости выбирать место

```python
preferred_seat = df.groupby('booking_complete')['wants_preferred_seat'].value_counts(normalize=True).to_frame().reset_index()
preferred_seat
```

```python
plt.figure(figsize=(5,5))
plt.subplot(1, 1, 1)
sns.barplot(x='booking_complete', y='proportion', hue = 'wants_preferred_seat', data=preferred_seat, errorbar=None)
```
