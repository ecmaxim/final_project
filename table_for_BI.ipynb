# Подготовка таблицы для визуализации в BI

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('alldata.csv')
df
```

```python
# меняем числовую категорию на текстовую, чтобы красиво отображалось в BI
df.loc[df['booking_complete'] == 0, 'booking_complete'] = 'Fail'
df.loc[df['booking_complete'] == 1, 'booking_complete'] = 'Success'
```

```python
df.booking_complete.value_counts()
```

```python
# Добавляем столбец pk-primary key
df['pk'] = pd.Series()
num=np.arange(1, 49282)
df['pk'] = num
df
```

```python
df.columns
```

```python
# корректируем порядок столбцов
df = df[['pk', 'num_passengers', 'sales_channel', 'trip_type', 'purchase_lead',
       'length_of_stay', 'flight_hour', 'flight_day', 'route',
       'booking_origin', 'wants_extra_baggage', 'wants_preferred_seat',
       'wants_in_flight_meals', 'flight_duration', 'booking_complete']]
```

```python
# сохраняем файл для использования в BI
df.to_csv(r"C:\Users\User\Python311_4\newclasswork\final_project\booking_for_BI.csv", index=False, sep=",")
```
